```python
# import libraries
import pandas as pd
import numpy as np

# Load data from excel

df1 = pd.read_excel("One2Car_list_Dec2020_page1-300.xlsx") 
df2 = pd.read_excel("One2Car_list_Dec2020_page301-600.xlsx") 
df3 = pd.read_excel("One2Car_list_Dec2020_page601-900.xlsx") 
df4 = pd.read_excel("One2Car_list_Dec2020_page901-1200.xlsx") 

# Mergeing the excel file togther 

df = pd.concat([df1,df1,df3,df4]) 
```

```python
df.info()
```

```python
df.describe()
```

```python
### Data cleaning section
## 1.Drop column unname
## 2.Mileage range and get rid of กม. 
## 3.Price get rid of บาท and เอาค่า \n\n   
## 4.Split column model_info to year,maker,model,engine_cc,sub_model,segment
```

```python
## 1.Drop column unnamed

df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
## 2.Mileage text modify

df['mileage(km)'] = df['mileage(km)'].str.replace(" ","")
df.head()
```

```python
# Delect กม. from column mileage(km)

df['mileage(km)'] = df['mileage(km)'].str.replace("กม.","")

df.head()
```

```python
# Replace K with "1000"

df['mileage(km)'] = df['mileage(km)'].str.replace("K","000")

df.head()
```

```python
# Split comlumn mileage(km) into mileage(km) and mileage(km)_mod

df['mileage(km)_mod'] = df['mileage(km)'].str.split('-').str[1]

df.head()
```

```python
# Fillna the value NaN with value from mileage(km)

df['mileage(km)_mod'].fillna(df['mileage(km)'], inplace=True)

df.head()
```

```python
# Drop mileage(km) column

df.drop(["mileage(km)"],axis=1,inplace=True)

df
```

```python
## Replace เกียร์อัตโนมัติ / เกียร์ธรรมดา with auto / manual

df['gear_type'] = df['gear_type'].replace('เกียร์อัตโนมัติ','auto')

df['gear_type'] = df['gear_type'].replace('เกียร์ธรรมดา','manual')

df
```

```python
# Modify price columns

df['Price'] = df['Price'].str.replace('บาท',' ')

df['Price(bath)'] = df['Price']

df.drop(['Price'],axis=1,inplace=True)

df
```

```python
# Delect , on price column

df['Price(bath)'] = df['Price(bath)'].str.replace(",","")

df
```

```python
# Change K in mileage(km)_mod column to 1000

df['mileage(km)_mod'] = df['mileage(km)_mod'].str.replace('k','000')

df['mileage(km)_mod'] = df['mileage(km)_mod'].str.replace(' ','')

df
```

```python
# Modify price more

df['Price'] = df['Price(bath)'].str.split().str[-1]

df['Price'] = df['Price'].str.replace("ติดต่อผู้ขาย","")

df
```

```python
# Delect column Price(bath) which is not clean

df.drop(['Price(bath)'],axis=1,inplace=True)

df
```

```python
# Extract year from model_info

df['year'] = df['model_info'].apply(lambda x: x.split(" ")[0])

df['maker'] = df['model_info'].apply(lambda x: x.split(" ")[1])

df['model'] = df['model_info'].apply(lambda x: x.split(" ")[2:3])

df['engineCC'] = df['model_info'].apply(lambda x: x.split(" ")[3:7])

df
```

```python
df.to_excel('clean_datadec2020.xlsx')
```
